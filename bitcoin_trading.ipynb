```python
#importing libraries

#gym libraies
import gym 
import gym_anytrading

#stable baseline libraries
from stable_baselines.common.vec_env import DummyVecEnv 
from stable_baselines import A2C

#basic Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bitcoin_data=pd.read_csv('gmedata.csv')
bitcoin_data.head()
```

```python
#creating the environment window_size takes first 5 signal_features
env=gym.make("stocks-v0",df=bitcoin_data,frame_bound=(5,100),window_size=5)
```

```python
env.prices
```

```python
env.signal_features
```

```python
env.action_space
```

```python
state = env.reset() #resetting the initial environment
while True: 
    action = env.action_space.sample() #took  random space , here action_space is 2 i.e. 0 and 1 which is buy and sell
    n_state, reward, done, info = env.step(action) #In env.step we are passing action i.e. buying or selling
    # n_state means what the environment looks like
    # reward means whether we made any profit 
    # done whether we have gone through every steps or not
    if done: 
        print("info", info) 
        break
        
plt.figure(figsize=(15,8))
plt.cla()
env.render_all()# showing the env of all data
plt.show()
```

```python
#building the training environment
env_maker= lambda: gym.make('stocks-v0',df=bitcoin_data, frame_bound=(5,100) ,window_size=5)
env=DummyVecEnv([env_maker])
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
model = A2C('MlpLstmPolicy', env, verbose=1) 
# verbose means  the information for training the model
model.learn(total_timesteps=1000000)
```
